### Concat

- concatanates pandas objects along an axis (rows, columns)
- there are set logics

```python
df1, df2
```

```python
pd.concat([df1, df2], axis=0) # adds rows as default, axis 0
```

```python
pd.concat([df1, df2], axis=1)
```

```python
df1, df3
```

```python
# concanates in axis rows, df1 rows first then df2 rows. Missing values are filled with NaNs
pd.concat([df1, df3], axis=1)

# = Outer Join: Allt från df1, allt från df3
```

## Merging

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K0", "K1", "K2"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)

left, right
```

```python
#syntax with function
# inner join -> K3 from right not included
pd.merge(left, right, on="key", indicator=True)
```

```python
# syntax with method
left.merge(right, on="key", how="outer", indicator=True)
```

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "A": ["A0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)
left, right
```

```python
# A_x is from the left, A_y is from the right
left.merge(right, on=["key1", "key2"], indicator=True, how="left")
```

## Join

- uses merge internally
- combines columns

```python
left = pd.DataFrame(
    {"A": ["A0", "A1", "A2"], "B": ["B0", "B1", "B2"]}, index=["K0", "K1", "K2"]
)


right = pd.DataFrame(
    {"C": ["C0", "C2", "C3"], "D": ["D0", "D2", "D3"]}, index=["K0", "K2", "K3"]
)

left, right
```

```python
# joins K1 which only exists in left -> so this is a left join
left.join(right)
```

```python
# indices from right -> right join
left.join(right, how="right")
```
